```python
import pandas as pd
import numpy as np
from itertools import chain
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df_2020 = pd.read_csv("C:/suhas/IT2020.csv")
```

```python
numeric_data = df_2020.select_dtypes(include=[np.number])
categorical_data = df_2020.select_dtypes(exclude=[np.number])

print('numeric dataset and ',numeric_data.shape)
print(' categorical dataset: ',categorical_data.shape )
```

```python
print(df_2020.City.value_counts())
```

```python
plt.boxplot(np.log10(df_2020['Yearly brutto salary (without bonus and stocks) in EUR']))
plt.xlabel('Annual salary without rewards')
plt.title('Yearly brutto salary (without bonus and stocks) in EUR')
plt.ylabel('log10(Salary)')
plt.grid()
plt.show()
```

```python
df_2020.groupby('Have you lost your job due to the coronavirus outbreak?').size().plot(kind='bar')
plt.xlabel('Jobs affected due to pandemic') 
plt.ylabel('Count') 
plt.title("People who lost job due to the coronavirus outbreak")
```

```python
sns.set_theme(style="darkgrid")
ax = sns.countplot(x="Employ",data=df_2020)

plt.xlabel('Ð¡ontract duration') 
plt.ylabel('Count') 
plt.title("Count of people who are having Different contracts")
```

```python
df_2020.Gender.str.get_dummies().sum().plot.pie(label='', autopct='%1.0f%%')
plt.title("Proportion of Genders")
```

```python
sns.set()
p8= plt.hist(df_2020['Age'],bins=10)
plt.xlabel('Age') 
plt.ylabel('Count') 
plt.title("Count of different Age")
plt.show(p8)
```

```python
ax = sns.boxplot(x="Age", y="Gender", data=df_2020)
plt.xlabel('Age') 
plt.ylabel('Gender') 
plt.title("Box Plot of Gender with diffrent Age Groups")
```

```python
sns.set_theme(style="darkgrid")
ax = sns.countplot(x="Company size",data=df_2020)
plt.xlabel('Company size') 
plt.ylabel('count') 
plt.title("Number of Employees in a Company")
```
