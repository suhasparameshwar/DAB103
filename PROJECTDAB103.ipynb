```python
import pandas as pd
import numpy as np
from itertools import chain
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns

```

# Loading the Data

```python
df_2020 = pd.read_csv("C:/suhas/IT2020.csv")
```

# Describing the Data

```python
df_2020
```

# Identifying The Data Types

```python
dataTypeSeries = df_2020.dtypes

print(dataTypeSeries)
```

# Visualization

# PLOT-1 (Checking outliers)

```python
plt.boxplot(np.log10(df_2020['Yearly brutto salary (without bonus and stocks) in EUR']))
plt.xlabel('Annual salary without rewards')
plt.title('YEARLY BRUTTO SALARY (without bonus and stocks) IN EUR (BOXPLOT)')
plt.ylabel('log10(Salary)')
plt.grid()
plt.show()
```

# Plot-2 

```python
df_2020.groupby('Have you lost your job due to the coronavirus outbreak?').size().plot(kind='bar') 
plt.xlabel('Jobs affected due to pandemic') 
plt.ylabel('Count') 
plt.title("EMPLOYEES WHO LOST JOB DUE TO CORONAVIRUS OUTBREAK")
```

# Plot-3

```python
sns.set()
ax = sns.countplot(x="Сontract duration",data=df_2020)

plt.xlabel('Сontract duration') 
plt.ylabel('Count') 
plt.title("COUNT OF PEOPLE WHO ARE HAVING DIFFERENT CONTRACTS")
```

# PLOT-4

```python

df_2020.Gender.str.get_dummies().sum().plot.pie(label='', autopct='%1.0f%%')
plt.title("PROPORTION OF GENDERS")
```

# PLOT-5

```python
sns.set_theme(style="darkgrid")
p8= plt.hist(df_2020['Age'],bins=10)
plt.xlabel('Age') 
plt.ylabel('Count') 
plt.title("COUNT OF DIFFERENT AGE")
plt.show(p8)
```

# PLOT-6

```python
sns.set_theme(style="darkgrid")
ax = sns.boxplot(x="Age", y="Gender", data=df_2020)
plt.xlabel('Age') 
plt.ylabel('Gender') 
plt.title("BOX PLOT OF GENDERS WITH RESPECTIVE TO THEIR AGES")
```

# PLOT-7

```python
sns.set_theme(style="darkgrid")
ax = sns.countplot(x="Company size",data=df_2020)
plt.xlabel('Company size') 
plt.ylabel('count') 
plt.title("DIFFERENT COMPANY SIZES OF PEOPLE AND ITS COUNT")
```
